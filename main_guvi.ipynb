# Handling Skewness

```python
import numpy as np
```

```python
df_new['transformed_price'] = np.sqrt(df_new['price']) #we use square root transformation as it is ideal for moderate skew
sns.displot(df_new['transformed_price'], kde = True)
plt.legend(labels = ['Skewness: %.2f'%(df_new['transformed_price'].skew())])
```

```python
df_new['transformed_num_subscribers'] = np.log(df_new['num_subscribers'] + 1)
sns.displot(df_new['transformed_num_subscribers'], kde = True)
plt.legend(labels = ['Skewness: %.2f' %(df_new['transformed_num_subscribers'].skew())])
```

# Building Models



# splitting datasets into training and testing

```python
from sklearn.model_selection import train_test_split
y = df_final['Rating']
x = df_final.drop('Rating', axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipeline = Pipeline([('scaler', StandardScaler()),('model', LinearRegression())])

pipeline.fit(x_train, y_train)

y_pred = pipeline.predict(x_test)
score1 = metrics.r2_score(y_test, y_pred)
score2 = metrics.mean_squared_error(y_test, y_pred)
score3 = metrics.mean_absolute_error(y_test, y_pred)
print('r2_score:',score1)
print('Mean Squared Error:', score2)
print('Mean Absolute Error:', score3)
```

```python
ax1 = sns.kdeplot(y_pred, label = 'predicted')
sns.kdeplot(y_test, label = 'actual', ax = ax1)
plt.xlim(0,1)
plt.legend()
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 3)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
pipeline = Pipeline([('scaler', StandardScaler()),('model', LinearRegression())])
pipeline.fit(x_train_poly, y_train)
y_pred = pipeline.predict(x_test_poly)
score1 = metrics.r2_score(y_test, y_pred)
score2 = metrics.mean_squared_error(y_test, y_pred)
score3 = metrics.mean_absolute_error(y_test, y_pred)
print('r2_score:',score1)
print('Mean Squared Error:', score2)
print('Mean Absolute Error:', score3)
```

```python
ax1 = sns.kdeplot(y_pred, label = 'predicted')
sns.kdeplot(y_test, label = 'actual', ax = ax1)
plt.xlim(0,1)
plt.legend()
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
pipeline = Pipeline([('scaler', StandardScaler()),
                     ('model', DecisionTreeRegressor(random_state =1, max_depth = 5, min_samples_split = 15,
                                                    min_samples_leaf = 5, max_features = 'log2'))])
pipeline.fit(x_train, y_train)
y_pred = pipeline.predict(x_test)
score1 = metrics.r2_score(y_test, y_pred)
score2 = metrics.mean_squared_error(y_test, y_pred)
score3 = metrics.mean_absolute_error(y_test, y_pred)
train_score = pipeline.score(x_train, y_train)
test_score = pipeline.score(x_test, y_test)
print('r2_score:',score1)
print('Mean Squared Error:', score2)
print('Mean Absolute Error:', score3)
print('Training score:', train_score)
print('Testing score:', test_score)


```
